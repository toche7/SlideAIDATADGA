```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')
```

```python
# Option A (recommended): load directly from GitHub
url = 'https://raw.githubusercontent.com/toche7/DataSets/main/president_heights.csv'
df = pd.read_csv(url)

# Option B: if you uploaded the file to Colab, use:
# df = pd.read_csv('/content/president_heights.csv')

df.head()
```

```python
print('Shape:', df.shape)
print('Columns:', list(df.columns))
print('\nMissing values by column:')
print(df.isna().sum())
print('\nData types:')
print(df.dtypes)
```

```python
height_col = 'height(cm)'

mean_height = df[height_col].mean()
max_height = df[height_col].max()
min_height = df[height_col].min()

tallest = df.loc[df[height_col].idxmax(), 'name']
shortest = df.loc[df[height_col].idxmin(), 'name']

print(f'Average height: {mean_height:.2f} cm')
print(f'Max height: {max_height:.0f} cm ({tallest})')
print(f'Min height: {min_height:.0f} cm ({shortest})')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='height(cm)', bins=10, kde=True, color='#2a9d8f')
plt.title('Distribution of U.S. Presidents Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
